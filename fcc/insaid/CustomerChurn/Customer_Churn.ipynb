```python
# Remove the duplicated rows in the dataframe
df_updated = df[df.duplicated()==False]
len(df_updated)
del df
```

```python
# Identifying and Handling missing data
missing_frame = pd.DataFrame(index=df_updated.columns.values)
missing_frame['Null Frequency'] = df_updated.isnull().sum().values
nullpercent = df_updated.isnull().sum().values/df_updated.shape[0]
missing_frame['Missing Null %age'] = np.round(nullpercent, decimals=4) *100
missing_frame.transpose()
```

### KNN Imputation
- Here data imputationis done by prediciton of a model

\* Imputation = Substituting Alternate data for missing values

```python
from sklearn.impute import KNNImputer
COLUMNS = ['TotalMorMin', 'TotalNightMin', 'TotalEveMin']
knn_imputer = KNNImputer()
raw_num = knn_imputer.fit_transform(df[COLUMNS])
raw_frame = pd.DataFrame(data=raw_num, columns=COLUMNS)
df['TotalMorMin'] = raw_frame['TotalMorMin'] 
df['TotalEveMin'] = raw_frame['TotalEveMin']
df['TotalNightMin'] = raw_frame['TotalNightMin']
```
