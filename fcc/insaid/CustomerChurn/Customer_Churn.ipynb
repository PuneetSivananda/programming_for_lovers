### KNN Imputation
- Here data imputationis done by prediciton of a model

\* Imputation = Substituting Alternate data for missing values

```python
from sklearn.impute import KNNImputer
COLUMNS = ['TotalMorMin', 'TotalNightMin', 'TotalEveMin']
knn_imputer = KNNImputer()
raw_num = knn_imputer.fit_transform(df[COLUMNS])
raw_frame = pd.DataFrame(data=raw_num, columns=COLUMNS)
df['TotalMorMin'] = raw_frame['TotalMorMin']
df['TotalEveMin'] = raw_frame['TotalEveMin']
df['TotalNightMin'] = raw_frame['TotalNightMin']
```

### Identification and handling Redundant Data

- create a custom function that handles duplicate features

```python
def duplicate_cols(data):
  ls1 = []
  ls2 = []
  columns =  data.columns.values
  for i in range(0, len(columns)):
    for j in range(i+1, len(columns)):
      if(np.where(data[columns[i]] == data[columns[j]], True, False).all()==True):
        ls1.append(columns[i])
        ls2.append(columns[j])

  if((len(ls1) == 0)&(len(ls2) == 0)):
    return None
  else:
    duplicate_frame = pd.DataFrame()
    duplicate_frame['Feature_1'] = ls1
    duplicate_frame['Feature_2'] = ls2
    return duplicate_frame
```

```python
print('Contains Redundant Records?', df.duplicated().any())
print('Diplicate Count:', df.duplicated().sum())
print('=====================================================')
print('Contains Redundant Features?', duplicate_cols(df))
```

### Performing Operations

```python
before_shape = df.shape
print("Data Shape [Before]", before_shape)

df.drop_duplicates(inplace = True)

after_shape = df.shape
print("Data Shape [After]", after_shape)

drop_nums = before_shape[0] - after_shape[0]
drop_percentage = np.round(drop_nums/before_shape[0], decimals=3) * 100

print("Drop Ratio: ", drop_percentage, '%')
```
